```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv('/content/Au_nanoparticle_dataset.csv')
```

```python
df
```

```python
new_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
```

```python
print(new_df.head(20))
```

```python
stats = new_df.describe()
print(stats.loc[['mean', 'std', '25%', '50%', '75%']])
```

```python
plt.figure(figsize=(20, 5))
for i, feature in enumerate(['N_total', 'N_bulk', 'N_surface', 'R_avg']):
    plt.subplot(1, 4, i+1)
    sns.histplot(new_df[feature], kde=True)
    plt.title(f'Histogram of {feature}')
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(new_df)
plt.show()
```

```python
g = sns.PairGrid(new_df)
g.map_upper(sns.histplot, bins=20)  # bivariate histogram
g.map_diag(sns.histplot, kde=True)  # histogram with KDE plot
g.map_lower(sns.kdeplot, cmap="Blues_d")  # bivariate KDE plot
plt.show()
```
